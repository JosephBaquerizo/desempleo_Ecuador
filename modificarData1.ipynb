# Obtención de datos
Se obtuvo el conjunto de datos y su respectivo diccionario a partir de el Portal Web de Datos Abiertos del Ecuador. El link es el siguiente: http://catalogo.datosabiertos.gob.ec/dataset/encuesta-nacional-de-empleo-desempleo-y-subempleo-enemdu-inec

Con la ayuda de Python, Excel y el diccionario adjuntado con los recursos, se pudo lograr una mejor comprensión de los datos que se estaban abarcando y determinar cómo era la estructura principal. 

En cada fila, los datos correspondientes a un feature determinado se encontraban separados por medios de puntos (.). No todos los valores estaban con datos incluidos, por lo que se mostraían 2 puntos seguidos indicando que dicha área de información no fue llenada en la Encuesta de Empleo, Desempleo y Subempleo.

Además, las filas no poseían una cantidad uniforme de comas (,). La cantidad de este caractér por cada fila podía variar entre 2 y 7 comas (por lo que si lo llevabamos a un formato de lista tendrían longitudes diferentes.

La cantidad de puntos existentes por cada fila tendía a permanecer en 191 puntos, sin embargo, en un feature específico del conjunto de datos, la información se encontraba entre comillas y podían existir más de 1 punto dentro de ese intervalo de caractéres (este intervalo sólo pertenecía a dicha área del feature). Por ello, a veces variaba el número de puntos existentes en cada fila, por lo que no se podría realizar un buen split de los datos. Por ejemplo, en esa área podría existir información del siguiente tipo: "Trabajad. amo de casa. obrero" (el cual posee dos puntos que no se deben tomar en cuenta al momento de separar los datos).

Debido a esto, se debía separar las columnas por puntos, pero era primordial obviar los puntos dentro del área de información indicada previamente para lograr una buena segmentación de datos.

Luego, se tuvo que obviar las comas dentro de cada fila (no proporcionaban una estructura importante en los datos y dificultaba el merge de los mismos) para obtener así un solo string por cada fila de datos y realizar las operaciones pertinentes según sea el caso.

```python
import csv
```

```python
lista = []

with open('bdd-enemdu-20130315anios.csv', 'r') as file:
    csv_reader = csv.reader(file)
    
    for line in csv_reader:
        lista.append(line)
        break
            
    
    next(csv_reader)
    
    
    i = 0
    for line in csv_reader:
        for elemento in line:
            if '\"' in elemento:
                indiceElemento = line.index(elemento)
                posComillas = []
                verificador = 0
                for letra in range(len(elemento)):
                    if elemento[letra] == '\"':
                        posComillas.append(letra)
                        verificador += 1
                    if verificador == 2:
                        elemento = elemento[0:posComillas[0]] + elemento[posComillas[0]:posComillas[1]+1].replace('.', ' ') + elemento[posComillas[1]+1:]
                        line[indiceElemento] = elemento
                        verificador = 3
        lista.append(line)
```

Se lee el archivo csv de la página de Datos Abiertos de Ecuador, se agregan los headers de las columnas y luego por cada elemento del archivo se trata de verificar si existen comillas y obtenemos la posición de dichas comillas. Reemplazamos el mismo texto pero con un space en vez de un punto y agregamos los elementos a la lista.

```python
with open('data1-bdd-enemdu-20130315anios.csv', 'w') as f:
    thewriter = csv.writer(f)
    
    for i in lista:
        thewriter.writerow(i)
```
