# Análisis de Empleo, Desempleo y Subempleo en el Ecuador
Se va a realizar un análisis sobre los factores que podrían tener una influencia sobre la situación laboral de las personas en el Ecuador a partir de varios registros obtenidos de el portal web Datos Abiertos de Ecuador.

Para esto es importante definir las situaciones laborales (que son 3):

    Empleo: Trabajo realizado por una persona, al cual se le atribuye una remuneración económica.
    Desempleo: Situación en la que una persona no se encuentra contratada para un empleo.
    Subempleo: Empleo realizado por un tiempo no completo, que por lo usual se remunera por debajo del mínimo salarial.
